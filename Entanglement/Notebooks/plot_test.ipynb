```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def count(x, y, N, p, beta0, theta, phi):
    return (N/4)*(p + (1-p)*(1+np.cos(2*y)*np.cos(2*x+beta0)-np.cos(2*theta)*(np.cos(2*y)+np.cos(2*x+beta0))+np.sin(2*theta)*np.sin(2*y)*np.sin(2*x+beta0)*np.cos(phi)))

beta = np.arange(0, 360, 10)
plt.plot(beta, count(beta*(np.pi/180), beta*(np.pi/180), N=2107.844	, p=0.049, beta0=0.121, theta=0.753, phi=0))
```

```python
def neg_normal(beta, mu, sigma):
    return -(1.0/((2.0*np.pi)**0.5*sigma))*(np.exp(-0.5*((beta-mu)/sigma)**2))*(250)

plt.plot(beta, neg_normal(beta*(np.pi/180), mu=180*(np.pi/180), sigma=1))
```

```python
def sinc(sai):
    return -np.sinc(sai)

beta = np.arange(0.1, 1000, 10)
beta1 = np.arange(-1000, 1000, 10)
plt.plot(beta1, sinc(beta1*np.pi/180)*count(beta1*(np.pi/180), beta1*(np.pi/180), N=2107.844, p=0.049, beta0=0.121, theta=0.753, phi=0))
```
