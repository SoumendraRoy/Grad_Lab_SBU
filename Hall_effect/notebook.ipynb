```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

# 300 K

```python
B, VH = np.loadtxt('hall_measure1.txt', skiprows=2, usecols=[0,1], unpack=True)
B_err = 0.01
VH_err = 0.001

plt.errorbar(B, VH, xerr=VH_err, yerr=B_err, fmt='.', label='300K')
plt.xlabel('Magnetic Field (mT)')
plt.ylabel('Hall Voltage (mV)')
plt.legend()
```

As the errorbars are constant (least count error), least square gives correct MLE.

```python
def y(B, a, b):
    return a*B + b

guess = [2.0/160.0, -1.0]

popt, pcov = curve_fit(y, B, VH, p0=guess)

plt.errorbar(B, VH, xerr=VH_err, yerr=B_err, fmt='.', label='300K')
plt.plot(B, y(B, *popt), color='red', label='fit')
plt.xlabel('Magnetic Field (mT)')
plt.ylabel('Hall Voltage (mV)')
plt.title('300 K')
plt.legend()
```

```python
print(popt, pcov)
```

```python
VR, I = np.loadtxt('vr_I.txt', skiprows=2, usecols=[0,1], unpack=True)

VR_err = 0.1
I_err = 0.001

plt.errorbar(VR, I, xerr=VR_err, yerr=I_err, fmt='.', label='300K')
plt.xlabel('Voltage across Sample (mV)')
plt.ylabel('Current (mA)')
plt.legend()
```

```python
def y1(VR, a1, b1):
    return a1*VR + b1

guess1 = [0.01/100.0, 0.0]

popt1, pcov1 = curve_fit(y1, VR, I, p0=guess1)

plt.errorbar(VR, I, xerr=VR_err, yerr=I_err, fmt='.', label='300K')
plt.plot(VR, y1(VR, *popt1), color='red', label='fit')
plt.xlabel('Voltage across Sample (mV)')
plt.ylabel('Current (mA)')
plt.legend()
```

```python
print(popt1, pcov1)
```
